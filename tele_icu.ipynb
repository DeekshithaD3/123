```python
import cv2
import os
from multiprocessing import Pool

# Function to extract frames from a single video
def extract_frames_from_video(video_path, output_dir, max_frames=100):
    cap = cv2.VideoCapture(video_path)
    frame_count = 0
    success = True

    while success and frame_count < max_frames:
        success, frame = cap.read()
        if success:
            frame_filename = os.path.join(output_dir, f"frame_{frame_count:05d}.jpg")
            cv2.imwrite(frame_filename, frame)
            frame_count += 1

    cap.release()

# Function to process each video
def process_video(video_file):
    video_path = os.path.join(video_dir, video_file)
    video_name = os.path.splitext(video_file)[0]
    video_frames_dir = os.path.join(frames_dir, video_name)
    os.makedirs(video_frames_dir, exist_ok=True)
    extract_frames_from_video(video_path, video_frames_dir, max_frames=100)  # Adjust max_frames as needed

# Define the directory containing the videos
video_dir = "/content/proj"
# Define the directory to save the frames
frames_dir = "/content/frames"

# Create the frames directory if it doesn't exist
os.makedirs(frames_dir, exist_ok=True)

# List all videos in the video directory
video_files = [f for f in os.listdir(video_dir) if f.endswith(".mp4")]

# Function to save progress regularly
def save_progress():
    for root, dirs, files in os.walk(frames_dir):
        level = root.replace(frames_dir, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print('{}{}/'.format(indent, os.path.basename(root)))
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            print('{}{}'.format(subindent, f))

# Use multiprocessing to process videos concurrently
with Pool(processes=os.cpu_count()) as pool:
    for i, _ in enumerate(pool.imap_unordered(process_video, video_files), 1):
        if i % 5 == 0:  # Save progress every 5 videos
            save_progress()

# Final save
save_progress()
```

#Define Dataset Structure

```python
import os
import random
import shutil

# Define directories
frames_dir = "/content/frames"
dataset_dir = "/content/drive/MyDrive/Colab Datasets/dataset"
train_dir = os.path.join(dataset_dir, "train")
val_dir = os.path.join(dataset_dir, "val")
test_dir = os.path.join(dataset_dir, "test")

# Create dataset directories if they don't exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
```

#List Video Folders

```python
# List video folders in frames_dir
video_folders = [folder for folder in os.listdir(frames_dir) if os.path.isdir(os.path.join(frames_dir, folder))]

# Shuffle and split dataset into train, validation, and test sets
random.seed(42)  # For reproducibility
random.shuffle(video_folders)

num_videos = len(video_folders)
num_train = int(0.7 * num_videos)
num_val = int(0.2 * num_videos)
num_test = num_videos - num_train - num_val

train_videos = video_folders[:num_train]
val_videos = video_folders[num_train:num_train + num_val]
test_videos = video_folders[num_train + num_val:]

print(f"Total videos: {num_videos}")
print(f"Train videos: {len(train_videos)}")
print(f"Validation videos: {len(val_videos)}")
print(f"Test videos: {len(test_videos)}")
```

#Move Frames to Dataset Folders

```python
def move_frames(video_list, dest_dir):
    for video_name in video_list:
        video_frames_dir = os.path.join(frames_dir, video_name)
        dest_video_frames_dir = os.path.join(dest_dir, video_name)
        os.makedirs(dest_video_frames_dir, exist_ok=True)
        # Move frames from frames_dir/video_name to dest_dir/video_name
        for frame_file in os.listdir(video_frames_dir):
            src = os.path.join(video_frames_dir, frame_file)
            dst = os.path.join(dest_video_frames_dir, frame_file)
            shutil.move(src, dst)

# Move frames to train, validation, and test directories
move_frames(train_videos, train_dir)
move_frames(val_videos, val_dir)
move_frames(test_videos, test_dir)

print("Dataset preparation complete.")
```

```python
import os
import cv2

def preprocess_images(dataset_path, preprocessed_path):
    os.makedirs(preprocessed_path, exist_ok=True)

    for root, _, files in os.walk(dataset_path):
        for file in files:
            if file.endswith(('.jpg', '.png')):
                img_path = os.path.join(root, file)
                img = cv2.imread(img_path)

                # Example preprocessing: resizing to 224x224
                img = cv2.resize(img, (224, 224))

                # Construct the save path relative to preprocessed_path
                relative_path = os.path.relpath(img_path, dataset_path)
                save_dir = os.path.join(preprocessed_path, os.path.dirname(relative_path))
                os.makedirs(save_dir, exist_ok=True)
                save_path = os.path.join(save_dir, file)

                cv2.imwrite(save_path, img)
                print(f'Preprocessed and saved: {save_path}')

    print('Preprocessing complete.')

preprocess_images(dataset_path, preprocessed_path)
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
dataset_path = '/content/drive/MyDrive/Colab Datasets/dataset'
preprocesed_path = '/content/drive/MyDrive/preprocessed_dataset'
```

```python
# Data preparation
def load_data(preprocessed_path, img_size=(224, 224)):
    datagen = ImageDataGenerator(rescale=0.255, validation_split=0.2)  # Normalize images and split for validation

    train_generator = datagen.flow_from_directory(
        preprocessed_path,
        target_size=img_size,
        batch_size=32,
        class_mode='categorical',
        subset='training'
    )

    validation_generator = datagen.flow_from_directory(
        preprocessed_path,
        target_size=img_size,
        batch_size=32,
        class_mode='categorical',
        subset='validation'
    )

    return train_generator, validation_generator

train_generator, validation_generator = load_data(preprocessed_path)
```

```python
# Build the model
def build_model(input_shape=(224, 224, 3), num_classes=2):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])

    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    return model

model = build_model(num_classes=train_generator.num_classes)
```

```python
# Train the model
epochs = 30
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=epochs
)
```

```python
model.save('object_detection_model.h5')

print('Model training complete and saved as object_detection_model.h5.')
```
